```python
#import all required libraries 

from __future__ import absolute_import, print_function
import os
import math 
from collections import Counter
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# read data and assign column names 
cols = (
   'date', 'flow_start', 'duration', 'protocol', 'src_addrIP',
    'direction', 'dst_addrIP', 'flags', 'tos', 'packets',
    'bytes', 'flows', 'label', 
)
dataset1 = pd.read_csv(
    'capture20110818.pcap.netflow.labeled', header=0,names=cols,skiprows=1, sep='\s+'
)
```

```python
dataset1
```

```python
dataset1['protocol'].replace("TCP", "6", regex=True, inplace=True)
dataset1['protocol'].replace("UDP", "17", regex=True, inplace=True)
dataset1['protocol'].replace("ICMP", "1", regex=True, inplace=True)
dataset1['protocol'].replace("PIM", "103", regex=True, inplace=True)
dataset1['protocol'].replace("RTP", "28", regex=True, inplace=True)
dataset1['protocol'].replace("ARP", "54", regex=True, inplace=True)
dataset1['protocol'].replace("IPX/SPX", "111", regex=True, inplace=True)
dataset1['protocol'].replace("R6", "6", regex=True, inplace=True)
dataset1['protocol'].replace("IGMP", "2", regex=True, inplace=True)
dataset1['protocol'].replace("IPV6-1", "43", regex=True, inplace=True)
dataset1['protocol'].replace("IPV6", "41", regex=True, inplace=True)
dataset1['protocol'].replace("ESP", "50", regex=True, inplace=True)
dataset1['protocol'].replace("LLC", "145", regex=True, inplace=True)
dataset1['protocol'].replace("UDT", "146", regex=True, inplace=True)
```

```python
dataset1['label'].replace("LEGITIMATE", "0", regex=True, inplace=True)
dataset1['label'].replace("Botnet", "1", regex=True, inplace=True)
dataset1['label'].replace("Background", "0", regex=True, inplace=True)
```

```python
dataset1['src_addr'], dataset1['src_port'] = dataset1['src_addrIP'].str.split(':',1).str
dataset1['dst_addr'], dataset1['dst_port'] = dataset1['dst_addrIP'].str.split(':',1).str
```

```python
dataset1
```

```python
dataset1 = dataset1.drop(columns=['date','flow_start','src_addrIP','direction','dst_addrIP','flags','src_addr','dst_addr'])
```

```python
dataset1
```

```python
dataset2 = dataset1
dataset2.duration = dataset2.duration.astype(float)
dataset2.protocol = pd.to_numeric(dataset2.protocol, downcast='float')
dataset2.tos = dataset2.tos.astype(float)
dataset2.packets = dataset2.packets.astype(float)
dataset2.bytes = dataset2.bytes.astype(float)
dataset2.flows = dataset2.flows.astype(float)
dataset2.label = dataset2.label.astype(float)
```

```python
dataset2 = dataset2.drop(columns=['src_port','dst_port'])
```

```python
y = dataset2.label
x = dataset2.drop('label', axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y,test_size=0.2)
```

```python
from sklearn import svm 
clf = svm.SVC()
clf.fit(X_train, y_train)
```

```python
y_score = clf.predict(X_test)
```

```python
from sklearn.metrics import average_precision_score
average_precision = average_precision_score(y_test, y_score)
print('Average precision-recall score: {0:0.2f}'.format(
      average_precision))
```
