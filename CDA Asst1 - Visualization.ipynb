```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
```

```python
a = pd.read_csv('data_for_student_case.csv')
# transactions_small["CreateDate"] = pd.to_datetime(transactions_small["CreateDate"])
a

a = a[["cvcresponsecode","bin", "simple_journal"]]
green = a.loc[a['simple_journal'] == "Chargeback"]
green = green[["cvcresponsecode","bin"]]

red = a.loc[a['simple_journal'] == "Settled"]
red = red[["cvcresponsecode","bin"]]

yellow = a.loc[a['simple_journal'] == "Refused"]
yellow = yellow[["cvcresponsecode","bin"]]



plt.scatter(red['bin'],red['cvcresponsecode'],c="red",marker="x",alpha=0.3) #settled
plt.scatter(green['bin'],green['cvcresponsecode'],c="green",marker=".") #chargeback
```

```python
plt.scatter(red['bin'],red['cvcresponsecode'],c="red",marker="x", alpha=0.2) #settled
plt.scatter(yellow['bin'],yellow['cvcresponsecode'],c="yellow",marker=".", alpha=0.09) #refused
plt.scatter(green['bin'],green['cvcresponsecode'],c="green",marker=".",alpha=0.1) #chargeback


```
